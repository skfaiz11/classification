##convert the categorical data to numercal data  using map function

```python
df['classification'] = df['classification'].map({'ckd':1,'notckd':0})
df['htn'] = df['htn'].map({'yes':1,'no':0})
df['rbc'] = df['rbc'].map({'abnormal':1,'normal':0})
df['pc'] = df['pc'].map({'abnormal':1,'normal':0})
df['pe'] = df['pe'].map({'yes':1,'no':0})
df['ane'] = df['ane'].map({'yes':1,'no':0})
df['dm'] = df['dm'].map({'yes':1,'no':0})



```

##creating X and Y

```python
X=df.values[:,0:-1]
Y=df.values[:,-1]
```

##Scaling the Data

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
sc.fit(X)
x=sc.transform(X)
print(X)
```

```python
Y=Y.astype(int)
```

##Splitting the Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.50, random_state = 42)
```

```python
# here our data is very less so i can split them into 50 training and 50 into testing
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# create a model
classifier= LogisticRegression() 
```

```python
#fitting training data to the model
classifier.fit(X_train, Y_train)
```

```python
#predict using the model
Y_pred = classifier.predict(X_test)  
```

```python
classifier.score(X_train,Y_train)
```

##Evaluating the Model

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report   # Model Evaluation 
```

```python
cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print('classification report: ')
print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print('accuracy of the model: ',acc)
```

##Adjusting the threshold

```python
#store the predicted probbilities
y_pred_prob=classifier.predict_proba(X_test)
print(y_pred_prob)
```

```python
Y_pred_class=[]
for value in y_pred_prob[:,1]:
  if value >0.40:
    Y_pred_class.append(1)                   #for decresing the type 2 error
  else:
    Y_pred_class.append(0)
print(Y_pred_class)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report   # Model Evaluation 
cfm=confusion_matrix(Y_test,Y_pred_class)
print(cfm)

print('classification report: ')
print(classification_report(Y_test,Y_pred_class))

acc=accuracy_score(Y_test,Y_pred_class)
print('accuracy of the model: ',acc)
```
